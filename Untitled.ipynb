```python
test = val.take(220)
val = val.skip(220)
```

```python
train = dataset
print(len(train))
print(len(test))
print(len(val))
```

```python
train_ds = train.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val.cache().shuffle(1000).prefetch(buffer_size = tf.data.AUTOTUNE)
```

```python
rescale_resize = tf.keras.Sequential([
    tf.keras.layers.experimental.preprocessing.Rescaling(1./255),
    tf.keras.layers.experimental.preprocessing.Resizing(120,120),
])
```

```python
cnn = tf.keras.Sequential([
    rescale_resize,
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3) , kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    tf.keras.layers.MaxPooling2D(pool_size=2 , strides=2),
    tf.keras.layers.Dropout(0.25),                      
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu' , kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),                      
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=1500, activation='relu',  input_dim=23040),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(38, activation='softmax') 
])
```

```python
cnn.build(input_shape=(None, 120, 120, 3))
cnn.summary()
```

```python
cnn.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),      
            loss= 'sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
training_history = cnn.fit(x = train_ds , validation_data= val_ds ,batch_size=8, epochs=10,callbacks=[early_stopping])
```

```python
cnn.save('prediction_plant1.h5')
print('save')
```
